Now let us analyse how Revenue is distributed country wise.  NOTE: This analysis is already done in Tableau.

```python
countrywise_total_revenue = battle.groupby('user_country').agg(total_revenue = ('user_cumulative_iap_revenue', 'sum'))
```

```python
countrywise_total_revenue
```

```python
largest_20_revenue_generators = countrywise_total_revenue.nlargest(20, columns = 'total_revenue', keep = 'first').reset_index()
largest_20_revenue_generators
```

```python
countries_highest_revenue = largest_20_revenue_generators['user_country'].to_list()
```

```python
countries_highest_revenue # list of countries which generate most amount of revenue
```

```python
countrywise_average_revenue = battle.groupby('user_country').agg(average_revenue = ('user_cumulative_iap_revenue', 'mean'))
largest_20_mean_revenue_generators = countrywise_average_revenue.nlargest(20, columns = 'average_revenue', keep = 'first').reset_index()
largest_20_mean_revenue_generators
```

```python
countries_highest_mean_revenue = largest_20_mean_revenue_generators['user_country'].to_list()
```

```python
countries_highest_mean_revenue # list of countries generating highest mean revenue
```

We can therefore focus on targeting the markets in the countries in both the lists!

```python
platformwise_average_revenue = battle.groupby('user_device_platform').agg(average_revenue = ('user_cumulative_iap_revenue', 'mean'))
largest_20_mean_revenue_generators = platformwise_average_revenue.nlargest(20, columns = 'average_revenue', keep = 'first').reset_index()
largest_20_mean_revenue_generators
```

```python
platformwise_total_revenue = battle.groupby('user_device_platform').agg(total_revenue = ('user_cumulative_iap_revenue', 'sum'))
largest_20_mean_revenue_generators = platformwise_total_revenue.nlargest(20, columns = 'total_revenue', keep = 'first').reset_index()
largest_20_mean_revenue_generators
```

The iOS platform is the biggest platform with regards to average revenue per user and total revenue platformwise.

```python
revenue_by_win = battle.groupby('event_battle_result').agg(avg_revenue= ('user_cumulative_iap_payments', 'mean'))
revenue_by_win
```

Conclusions: 

1. At the moment, there is greater scope for those with high `user_point_in_lifetime` to contribute to revenue. 
2. The mild correlation with `map_id` reveals that the revenue does not significantly change with the `map`. 
Perhaps lesser resources can be dedicated to building multiple maps unless there are a few outstanding ones
that do better on this metric. 
3. 2-star wins correspond to half of the revenue of the other categories. Perhaps more needs to be done to 
incentivise 2-star victors to conduct more in-game transactions.
